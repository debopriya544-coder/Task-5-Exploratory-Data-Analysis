```python
##Import libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Load csv dataset
```

```python
train=pd.read_csv("train.csv")
train
```

```python
## 1.Basic exploration
```

```python
print("=== TRAIN DATA INFO ===")
print(train.info())
print("\n=== TRAIN DATA DESCRIPTION ===")
print(train.describe())

print("\n=== VALUE COUNTS: Survived ===")
print(train['Survived'].value_counts())

print("\n=== VALUE COUNTS: Pclass ===")
print(train['Pclass'].value_counts())

print("\n=== VALUE COUNTS: Sex ===")
print(train['Sex'].value_counts())
```

```python
## 2.pairplot for numeric column
```

```python
sns.pairplot(train.select_dtypes(include=['float64', 'int64']), diag_kind='hist')
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
# Observation:
# Pairplot shows potential relationships between features like Age, Fare, and Survived.
# Age and Fare appear to have some distribution differences for survivors.
```

```python
## 3.Histograms
```

```python
train.hist(figsize=(12,8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Titanic Train Data", y=1.02)
plt.show()
```

```python
# Observation:
# Age distribution is right-skewed; most passengers are between 20-40 years.
# Fare distribution is highly skewed with some outliers.
```

```python
# 4.Boxplots
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Age', data=train)
plt.title("Age vs Survival")
plt.show()
```

```python
# Observation:
# Younger passengers appear to have slightly higher survival rates.
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Fare', data=train)
plt.title("Fare vs Survival")
plt.show()
```

```python
# Observation:
# Higher fares seem associated with higher survival rates.
```

```python
## 5.scatterplot
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train, palette='coolwarm')
plt.title("Age vs Fare colored by Survival")
plt.show()
```

```python
# Observation:
# Many survivors paid higher fares and were often in younger to middle age.
```

```python
# 6. Summary of Findings
```

```python
summary = """
Summary of Titanic Data Findings:
1. Class and fare have a strong relationship with survival: higher class & higher fare = better survival.
2. Women survived at a much higher rate than men (seen in value_counts).
3. Younger passengers tended to survive more often than older ones.
4. Age has weak correlation but still shows survival trends in boxplots.
5. The Fare feature is highly skewed; high fare outliers often survived.
"""
print(summary)
```
